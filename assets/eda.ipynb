## Model training

### MORE ENCODING UNFORTUNATELY

```python
data_ML = data1.drop("gpa_2023", axis=1)
data_ML.columns
```

```python
data_ML['parent_strength'].unique()
```

```python
data_ML.isna().sum()
```

```python
data_ML['parent_strength'].value_counts()
```

```python
data_ML.loc[data_ML["sex"]== "Male", "sex"] = 0
data_ML.loc[data_ML["sex"]== "Female", "sex"] = 1
data_ML['sex'] = data_ML['sex'].astype(int)

data_ML.loc[data_ML["study_year"]== "1st Year", "study_year"] = 0
data_ML.loc[data_ML["study_year"]== "2nd Year", "study_year"] = 1
data_ML.loc[data_ML["study_year"]== "3rd Year", "study_year"] = 2
data_ML.loc[data_ML["study_year"]== "4th Year", "study_year"] = 3
data_ML.loc[data_ML["study_year"]== "Postgraduate", "study_year"] = 4
data_ML['study_year'] = data_ML['study_year'].astype(int)

data_ML.loc[data_ML["faculty"]== "AgriSciences", "faculty"] = 0
data_ML.loc[data_ML["faculty"]== "Arts & Social Sciences", "faculty"] = 1
data_ML.loc[data_ML["faculty"]== "Economic & Management Sciences", "faculty"] = 2
data_ML.loc[data_ML["faculty"].isna(), "faculty"] = 2
data_ML.loc[data_ML["faculty"]== "Education", "faculty"] = 3
data_ML.loc[data_ML["faculty"]== "Engineering", "faculty"] = 4
data_ML.loc[data_ML["faculty"]== "Law", "faculty"] = 5
data_ML.loc[data_ML["faculty"]== "Medicine and Health Services", "faculty"] = 6
data_ML.loc[data_ML["faculty"]== "Military Science", "faculty"] = 7
data_ML.loc[data_ML["faculty"]== "Science", "faculty"] = 8
data_ML.loc[data_ML["faculty"]== "Theology", "faculty"] = 9
data_ML['faculty'] = data_ML['faculty'].astype(int)

data_ML.loc[data_ML["accomodation"]== "Private accommodation/ stay with family/friends", "accomodation"] = 0
data_ML.loc[data_ML["accomodation"].isna(), "accomodation"] = 0
data_ML.loc[data_ML["accomodation"]== "Non-private accommodation ie. Res", "accomodation"] = 1
data_ML['accomodation'] = data_ML['accomodation'].astype(int)

data_ML.loc[data_ML["allowance"]== "No Allowance", "allowance"] = 0
data_ML.loc[data_ML["allowance"]== "Ksh 30k - 35k", "allowance"] = 1
data_ML.loc[data_ML["allowance"]== "Ksh 36k - 42k", "allowance"] = 2
data_ML.loc[data_ML["allowance"]== "Ksh 43k - 50k", "allowance"] = 3
data_ML.loc[data_ML["allowance"]== "Ksh 51k - 57k", "allowance"] = 4
data_ML.loc[data_ML["allowance"]== "Ksh 57k+", "allowance"] = 5
data_ML['allowance'] = data_ML['allowance'].astype(int)

data_ML.loc[data_ML["scholarship"]== "No", "scholarship"] = 0
data_ML.loc[data_ML["scholarship"].isna(), "scholarship"] = 0
data_ML.loc[data_ML["scholarship"]== "Yes (NSFAS, etc...)", "scholarship"] = 1
data_ML['scholarship'] = data_ML['scholarship'].astype(int)

data_ML.loc[data_ML["weekly_study_hrs"]== "1-3", "weekly_study_hrs"] = 1
data_ML.loc[data_ML["weekly_study_hrs"]== "3-5", "weekly_study_hrs"] = 2
data_ML.loc[data_ML["weekly_study_hrs"]== "5-8", "weekly_study_hrs"] = 3
data_ML.loc[data_ML["weekly_study_hrs"]== "8+", "weekly_study_hrs"] = 4
data_ML['weekly_study_hrs'] = data_ML['weekly_study_hrs'].astype(int)

data_ML.loc[data_ML["socializing_freq"]== "4+", "socializing_freq"] = 4
data_ML.loc[data_ML["socializing_freq"]== "Only weekends", "socializing_freq"] = 5
data_ML['socializing_freq'] = data_ML['socializing_freq'].astype(int)

data_ML.loc[data_ML["drinks_per_night"]== "1-3", "drinks_per_night"] = 1
data_ML.loc[data_ML["drinks_per_night"]== "3-5", "drinks_per_night"] = 2
data_ML.loc[data_ML["drinks_per_night"]== "5-8", "drinks_per_night"] = 3
data_ML.loc[data_ML["drinks_per_night"]== "8+", "drinks_per_night"] = 4
data_ML['drinks_per_night'] = data_ML['drinks_per_night'].astype(int)

data_ML.loc[data_ML["missed_classes"]== "4+", "missed_classes"] = 4
data_ML.loc[data_ML["missed_classes"].isna(), "missed_classes"] = 0
data_ML['missed_classes'] = data_ML['missed_classes'].astype(int)

data_ML.loc[data_ML["failed_modules"]== "4+", "failed_modules"] = 4
data_ML.loc[data_ML["failed_modules"].isna(), "failed_modules"] = 0
data_ML['failed_modules'] = data_ML['failed_modules'].astype(int)

data_ML.loc[data_ML["relationship"]== "Yes", "relationship"] = 0
data_ML.loc[data_ML["relationship"].isna(), "relationship"] = 1
data_ML.loc[data_ML["relationship"]== "No", "relationship"] = 1
data_ML['relationship'] = data_ML['relationship'].astype(bool)

data_ML.loc[data_ML["approval"]== "Yes", "approval"] = 0
data_ML.loc[data_ML["approval"].isna(), "approval"] = 1
data_ML.loc[data_ML["approval"]== "No", "approval"] = 1
data_ML['approval'] = data_ML['approval'].astype(bool)

data_ML.loc[data_ML["parent_strength"]== "Very close", "parent_strength"] = 0
data_ML.loc[data_ML["parent_strength"].isna(), "parent_strength"] = 0
data_ML.loc[data_ML["parent_strength"]== "Close", "parent_strength"] = 1
data_ML.loc[data_ML["parent_strength"]== "Fair", "parent_strength"] = 2
data_ML.loc[data_ML["parent_strength"]== "Distant", "parent_strength"] = 3
data_ML['parent_strength'] = data_ML['parent_strength'].astype(int)

data_ML.info()
```

### Training

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
X = data_ML.drop('gpa', axis=1)
y = data_ML['gpa']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=1566)

knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```
